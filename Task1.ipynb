# **2. Інтуїція Логістичної Регресії** <a class="anchor" id="2"></a>


[Таблиця змісту](#0.1)


В статистиці, **модель Логістичної Регресії** — це широко використовувана статистична модель, що в основному використовується для цілей класифікації. Це означає, що при заданому набору спостережень, алгоритм Логістичної Регресії допомагає нам класифікувати ці спостереження на два або більше дискретних класів. Тож, цільова змінна має дискретний характер.


Алгоритм Логістичної Регресії працює наступним чином -

## **Реалізація лінійного рівняння**


Алгоритм Логістичної регресіїї працює шляхом реалізації лінійного рівняння з незалежними чи описовими змінними для передбачення значення відповіді. Для прикладу, розглянемо зразок кількості годин навчання та ймовірність складання іспит. Тут, кількість годин навчання є описовою змінною і позначається вона як x1. Ймовірність складання іспиту є змінною відповіді або цільовою змінною, і позначається вона як z.


Якщо ми маємо одну описову та одну змінну (x1) та одну відповідну змінну (z), тоді лінійне рівняння математично буде задано наступним рівнянням-

    z = β0 + β1x1    

Тут, коефіцієнти β0 та β1 є параметрами моделі.


Якщо описових змінних декілька, тоді наведене вище рівняння можна розширити до

    z = β0 + β1x1+ β2x2+……..+ βnxn
    
Тут, коефіцієнти β0, β1, β2 та βn є параметрами моделі.

Отже, прогнозоване значення відповіді задається наведеними вище рівняннями та позначається як z.

## **Сигмоїда**

Ця передбачена змінна відповіді, позначена як z, далі конвертується в значення ймовірності, що лежить між 0 та 1. Ми використовуємо сигмоїду для того, щоб перетворити прогнозовані значення на значення ймовірностей. Ця сигмоїда потім перетворює будь-яке дійсне значення на значення ймовірності між 0 та 1.

У машинному навчанні, сигмоїда використовується для перетворення передбачень на ймовірності. Сигмоїда функція має S-подібну криву. Її також називають сигмоподібною кривою.

Сигмоїда — це особливий випадок Логістичної функції. Вона задається наступною математичною формулою.

Графічно, ми можемо представити сигмоїду за допомогою наступного графіка.

## **Межа прийняття рішення**

Сигмоїда повертає значення ймовіртності між 0 та 1. Це значення ймовіртності потім переноситься до дискретного класу, який є “0” або “1”. Для того, щоб віднести це значення ймовіртності на дискретний клас (пройшов/провалився, так/ні, правда/неправда), ми обираємо порогове значення. Це порогове значення називається Межою прийняття рішення. Вище цього порогового значення, ми переносимо значення ймовірності до класу 1, а нижче — до класу 0.

Математично, це можна виразити наступним чином:-

p ≥ 0.5 => клас = 1

p < 0.5 => клас = 0 

Зазвичай, межа прийняття рішення встановлюється як 0.5. Тож, якщо значення ймовірності є 0.8 (> 0.5), ми перенесемо це спостереження до класу 1. Так само, якщо значення ймовірності є 0.2 (< 0.5), ми перенесемо це спостереження до класу 0. Це представлено на графіку нижче-

![Decision boundary in sigmoid function](https://ml-cheatsheet.readthedocs.io/en/latest/_images/logistic_regression_sigmoid_w_threshold.png)

## **Складання прогонзів**

Тепер, ми знаємо про сигмоїду та межу прийняття рішення в логістичній регресії. Ми можемо використати наші знання про сигмоїду та межу прийняття рішення, щоб написати функцію передбачення. Функція передбачення в логістичній регресії повертає ймовірність того, що спостереження є позиитвним, Так чи Правда. Ми називаємо це класом 1 і він позначається як P(class = 1). Якщо ймовірність наближається до одиниці, тоді ми будемо більш впевнені щодо нашої моделі, що спостереження відноситься до класу 1, інакше воно відноситься до класу 0

Модель Логістичної Регресії вимагає кількох ключових припущень. Це ось такі:-

1. Модель Логістичної Регресії вимагає, щоб залежна змінна мала бінарний, багаточленний або порядковий характер.

2. Вона вимагає, щоб спостереження були незалежними одне від одного. Отже, спостереження не мають виходити з повторних вимірювань.

3. Алгоритм Логістичної Регресії вимагає невеликої або повної відсутності мультиколінеарності між незалежними змінними. Це означає, що незалежні змінні не мають надто корелювати одна з іншою.

4. Модель Логістичної Регресії припускає лінійність незалежних змінних та логарифмічних шансів.

5. Успіх моделі Логістичної Регресії залежить від розмірів вибірки. Як правило, вимагається великий розмір вибірки, щоб досягти високої точності.

### Припущення


Я припускаю, що дані відсутні зовсім випадково (MCAR, missing completely at random). Є два методи, які можна використати для вписування відсутніх значень. Перший – це середня або медіанна імпутація, а інший – це імпутація випадкової вибірки. Коли в наборі даних є викиди, ми повинні використовувати медіанну імпутацію. Тож, я використовуватиму медіанну імпутацію, тому що вона стійка до викидів.


Я імпутую відсутні значення відповідними статистичними показниками даних, в цьому випадку – медіаною. імпутацію слід виконувати на тренувальному наборі, а потім поширювати на випробувальний набір. Це означає, що статистичні показники, які будуть використовуватися для заповнення відсутніх значень, як в тренувальному так і в випробувальному наборі, повинні бути витягнуті лише з тренувального. Це для того, щоб уникнути перенавчання.

```python
# імпутувати відсутні значення в X_train та X_test медіаною відповідного стовпця у X_train

for df1 in [X_train, X_test]:
    for col in numerical:
        col_median=X_train[col].median()
        df1[col].fillna(col_median, inplace=True)           
      
```

Оцінка точності тренувального набору становить 0.8476, тоді як точність випробувального набору – 0.8502. Ці два значення цілком порівнянні. Отже, про перенавчання мова не йдеться.


У Логістичній Регресії, ми використовуємо значення C = 1 за замовчуванням. Він забезпечує хорошу продуктивність із приблизно 85% точністю як на тренувальному, так і на випробувальному наборі. Але продуктивність моделі як на тренувальному, так і на випробувальному наборі дуже порівнянна. Ймовірно, справа в недонавчанні.

Я збільшу C та навчу більш гнучку модель.

```python
# навчити модель Логістичній Регресії із C=100

# створити екземпляр моделі
logreg100 = LogisticRegression(C=100, solver='liblinear', random_state=0)


# навчити модель
logreg100.fit(X_train, y_train)
```

```python
# надрукувати оцінки тренувального та тестового наборів

print('Оцінка тренувального набору: {:.4f}'.format(logreg100.score(X_train, y_train)))

print('Оцінка випробувального набору: {:.4f}'.format(logreg100.score(X_test, y_test)))
```

Можемо бачити, що C=100 призводить до вищої точності випробувального набору, а також дещо підвищеної точності тренувального набору. Отже, ми можемо зробити висновок, що більш складна модель має працювати краще.

Тепер я досліджу, що станеться, якщо ми використовуємо більш регуляризовану модель, ніж при значенні за замовчуванням C=1, встановивши C=0.01.

```python
# навчити модель Логістичній Регресії із C=0.01

# створити екземпляр моделі
logreg001 = LogisticRegression(C=0.01, solver='liblinear', random_state=0)


# навчити модель
logreg001.fit(X_train, y_train)
```

```python
# надрукувати оцінки тренувального та тестового наборів

print('Оцінка тренувального набору {:.4f}'.format(logreg001.score(X_train, y_train)))

print('Оцінка випробувального набору: {:.4f}'.format(logreg001.score(X_test, y_test)))
```

Отже, якщо ми використовуємо більш регуляризовану модель, встановивши C=0.01, то точність як тренувального, так і випробувального набору зменшується відносно параметрів за замовчуванням.

# **15. Матриця невідповідностей** <a class="anchor" id="15"></a>


[Таблиця змісту](#0.1)


Матриця невідповідностей — це інструмент для узагальнення продуктивності алгоритму класифікації. Матриця невідповідностей дасть нам чітку картину продуктивності моделі класифікації та типів помилок, створених цією моделлю. Вона дає нам зведення правильних і неправильних прогнозів, розбитих за кожною категорією. Зведення надані у табличному вигляді.


Можливі чотири типи результатів під час оцінки продуктивності класифікаційної моделі. Ці чотири результати описані нижче:-

**Істинно Позитивні (TP)** – Істинно Позитивні трапляються, коли ми передбачаємо, що спостереження належить до певного класу, і це спостереження насправді належить до того класу.


**Істинно Негативні (TN)** – Істинно Негативні трапляються, коли ми передбачаємо, що спостереження не належить до певного класу, і це спостереження насправді не належить до того класу.


**Хибно Позитивні (FP)** – Хибно Позитивні трапляються, коли ми передбачаємо, що спостереження належить до певного класу, але насправді це спостереження не належить до того класу. Цей тип помилки називається **Помилка I роду.**



**Хибно негативні (FN)** – Хибно негативні трапляються, коли ми передбачаємо, що спостереження не належить до певного класу, але насправді це спостереження належить до того класу. Ця помилка дуже серйозна і називається вона **Помилка II роду.**



Ці чотири результати зведені в матриці невідповідностей, наведеній нижче.


```python
# Роздрукувати Матрицю Невідповідностей та розділити її на чотири шматки

from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test, y_pred_test)

print('Матриця Невідповідностей\n\n', cm)

print('\nІстинно Позитивні(TP) = ', cm[0,0])

print('\nІстинно Негативні(TN) = ', cm[1,1])

print('\nХибно Позитивні(FP) = ', cm[0,1])

print('\nХибно Негативні(FN) = ', cm[1,0])
```

Матриця Невідповідностей показує `20892 + 3286 = 24178 правильних прогнозів` та `3086 + 1175 = 4261 неправильний прогноз`.


У цьому випадку, маємо


- `Істинно Позитивні` (Справді Позитивні:1 та Прогнозовані Позитивні:1) - 20892


- `Істинно Негативні` (Справді Негативні:0 та Прогнозовані Негативні:0) - 3286


- `Хибно Позитивні` (Справді Negative:0 але Прогнозовані Позитивні:1) - 1175 `(Помилка I роду)`


- `Хибно Негативні` (Справді Позитивні:1 але Прогнозовані Негативні:0) - 3086 `(Помилка II роду)`

```python
# візуалізувати матрицю невідповідностей за допомогою теплокарти seaborn

cm_matrix = pd.DataFrame(data=cm, columns=['Дійсні Позитивні:1', 'Дійсні Негативні:0'], 
                                 index=['Прогнозовані Позитивні:1', 'Прогнозовані Негативні:0'])

sns.heatmap(cm_matrix, annot=True, fmt='d', cmap='YlGnBu')
```

## Класифікаційний звіт


**Класифікаційний звіт** – це ще один спосіб оцінити продуктивність класифікаційної моделі. Він демонструє  оцінки **точності**, **відклику**, **f1** та **підтримки** для моделі. Ці терміни я описав далі.

Можемо надрукувати класифікаційний звіт наступним чином:-

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, y_pred_test))
```

## Точність класифікації

```python
TP = cm[0,0]
TN = cm[1,1]
FP = cm[0,1]
FN = cm[1,0]
```

```python
# надрукувати точність класифікації

classification_accuracy = (TP + TN) / float(TP + TN + FP + FN)

print('Точність класифікації : {0:0.4f}'.format(classification_accuracy))
```

## Помилка класифікації

```python
# надрукувати помилку класифікації

classification_error = (FP + FN) / float(TP + TN + FP + FN)

print('Помилка класифікації : {0:0.4f}'.format(classification_error))
```

## Точність


**Точність** можна визначити як відсоток правильно передбачених позитивних результатів із усіх прогнозованих позитивних результатів. Її можна подати як відношення Істинно Позитивних результатів (TP) до суми Істинно та Хибно Позитивних результатів (TP + FP).


Отже, **Точність** визначає частку правильно прогнозованого позитивного результату. Вона більше має справу із позитивним класом, ніж із негативним.



Математично, точність можна визначити як відношення `TP до (TP + FP).`




```python
# надрукувати оцінку точності

precision = TP / float(TP + FP)


print('Точність : {0:0.4f}'.format(precision))
```

## Відклик


Відклик можна визначити як відсоток правильно передбачених позитивних результатів від усіх фактичних позитивних результатів.
Його можна представити як відношення Істинно Позитивних (TP) до суми Істинно Позитивних та Хибно Негативних (TP + FN). **Відклик** також називається **Чутливістю**.


**Відклик** визначає частку правильно передбачених фактичних позитивних результатів.


Математично, відклик можна визначити як відношення `TP до (TP + FN).`





```python
recall = TP / float(TP + FN)

print('Відклик або Чутливість : {0:0.4f}'.format(recall))
```

## Істиннопозитивний Рівень


**Істиннопозитивний Рівень** є синонімом **Відклику**.


```python
true_positive_rate = TP / float(TP + FN)


print('Істиннопозитивний Рівень : {0:0.4f}'.format(true_positive_rate))
```

## Хибнопозитивний Рівень

```python
false_positive_rate = FP / float(FP + TN)


print('Хибнопозитивний Рівень : {0:0.4f}'.format(false_positive_rate))
```

## Специфічність

```python
specificity = TN / (TN + FP)

print('Специфічність : {0:0.4f}'.format(specificity))
```

## f1-оцінка


**f1-оцінка** – це зважене гармонічне середнє між точностю та відкликом. Найкраща можлива **f1-оцінка** буде 1.0, а найгірша – 0.0. **f1-оцінка** – це гармонічне середнє між точностю та відкликом. Отже, **f1-оцінка** є завжди нижчою за показники правильності, оскільки вони заклали точність та відклик у свої обчислення. Для порівняння моделей класифікаторів слід використовувати середнє зважене значення `f1-оцінки`, а не глобальну точність.



## Підтримка


**Підтримка** — це фактична кількість входжень класу в нашому наборі даних.

```python
from sklearn.preprocessing import binarize

for i in range(1,5):
    
    cm1=0
    
    y_pred1 = logreg.predict_proba(X_test)[:,1]
    
    y_pred1 = y_pred1.reshape(-1,1)
    
    y_pred2 = binarize(y_pred1, threshold = i/10)
    
    y_pred2 = np.where(y_pred2 == 1, 'Yes', 'No')
    
    cm1 = confusion_matrix(y_test, y_pred2)
        
    print ('Із',i/10,'порог Матриці Невідповідностей становить ','\n\n',cm1,'\n\n',
           
            'із',cm1[0,0]+cm1[1,1],'правильними прогнозами, ', '\n\n', 
           
            cm1[0,1],'помилок I роду (Хибно Позитивні), ','\n\n',
           
            cm1[1,0],'помилок II роду (Хибно Негативні), ','\n\n',
           
           'Оцінка точності: ', (accuracy_score(y_test, y_pred2)), '\n\n',
           
           'Чутливість: ',cm1[1,1]/(float(cm1[1,1]+cm1[1,0])), '\n\n',
           
           'Специфічність: ',cm1[0,0]/(float(cm1[0,0]+cm1[0,1])),'\n\n',
          
            '====================================================', '\n\n')
```

### Коментарі


- У бінарних задачах, за замовчуванням використовується поріг 0.5 для перетворення передбачених ймовірностей у класові передбачення.


- Пороговий рівень може бути налаштований для підвищення чутливості або специфічності. 


- Чутливість і специфічність мають зворотню пропорційність. Збільшення однієї величини завжди зменшує іншу і навпаки.


- Можем бачити, що збільшення порогового рівня призводить до підвищенної точності.


- Налаштування порогового рівня має бути одним із останніх кроків у процесі побудови моделі.

## ROC-крива


Ще один інструмент для візуального вимірювання ефективності моделі класифікації — це **ROC-крива**. ROC-крива означає **Receiver Operating Characteristic Curve (Крива Робочої Характеристики Приймача)**. **ROC-крива** – це графік, що показує ефективність класифікаційної моделі на різних порогових рівнях класифікації. 



**ROC-крива** будує **Істиннопозитивний Рівень (TPR)** напроти **Хибнопозитивного Рівня (FPR)** на різних порогових рівнях.



**Істиннопозитивний Рівень (TPR)** також називається **Відкликом**. Він визначається як відношення `TP до (TP + FN).`



**Хибнопозитивний Рівень (FPR)** визначається як відношення `FP до (FP + TN).`




В ROC-кривій, ми зосереджемось на TPR (Істиннопозитивний Рівень) та FPR (Хибнопозитивний Рівень) однієї точки. Це дасть нам загальну продуктивність ROC-кривої, яка складається з TPR і FPR на різних порогових рівнях. Отже, ROC-крива будує TPR проти FPR на різних порогових рівнях класифікації. Якщо ми знизимо порогові рівні, це може призвести до того, що елементів, класифікованих як позитивні, буде більше. Це збільшить як Істинно Позитивні (TP), так і Хибно Позитивні результати (FP).



```python
# побудувати ROC-криву

from sklearn.metrics import roc_curve

fpr, tpr, thresholds = roc_curve(y_test, y_pred1, pos_label = 'Yes')

plt.figure(figsize=(6,4))

plt.plot(fpr, tpr, linewidth=2)

plt.plot([0,1], [0,1], 'k--' )

plt.rcParams['font.size'] = 12

plt.title('ROC-крива для класифікатора RainTomorrow')

plt.xlabel('Хибнопозитивний Рівень (1 - Специфічність)')

plt.ylabel('Істиннопозитивний Рівень (Чутливість)')

plt.show()
```

ROC-крива допомагає нам вибрати пороговий рівень, який збалансовує чутливість та специфічність для певної ситуації.

## ROC-AUC


**ROC AUC** означає **Receiver Operating Characteristic - Area Under Curve (Робоча Характеристика Приймача - Площа Під Кривою)**. Це техніка для порівняння продуктивності класифікатора. У цій техніці ми вимірюємо `площу під кривою (AUC)`. Ідеальний класифікатор матиме ROC AUC дорівним 1, тоді як чисто випадковий класифікатор матиме ROC AUC дорівним 0.5. 


Тож, **ROC AUC** — це відсоток ROC-графіку, який знаходиться під кривою.

```python
# обчислити ROC AUC

from sklearn.metrics import roc_auc_score

ROC_AUC = roc_auc_score(y_test, y_pred1)

print('ROC AUC : {:.4f}'.format(ROC_AUC))
```

### Коментарі


- ROC AUC — це єдиний підсумковий показник продуктивності класифікатора. Чим вище значення, тим краще класифікатор.

- ROC AUC нашої моделі наближається до 1. Таким чином, ми можемо зробити висновок, що наш класифікатор добре справляється з прогнозуванням того, буде завтра дощ чи ні.

```python
# розрахувати перехресне затверджування ROC AUC 

from sklearn.model_selection import cross_val_score

Cross_validated_ROC_AUC = cross_val_score(logreg, X_train, y_train, cv=5, scoring='roc_auc').mean()

print('Перехресне затверджування ROC AUC : {:.4f}'.format(Cross_validated_ROC_AUC))
```

# **20. Оптимізація гіперпараметрів із використанням GridSearch CV** <a class="anchor" id="20"></a>


[Таблиця змісту](#0.1)

```python
from sklearn.model_selection import GridSearchCV


parameters = [{'penalty':['l1','l2']}, 
              {'C':[1, 10, 100, 1000]}]



grid_search = GridSearchCV(estimator = logreg,  
                           param_grid = parameters,
                           scoring = 'accuracy',
                           cv = 5,
                           verbose=0)


grid_search.fit(X_train, y_train)
```

```python
# дослідити найкращу модель

# найкращий результат, досягнутий під час GridSearchCV
print('Найкращий результат GridSearchCV : {:.4f}\n\n'.format(grid_search.best_score_))

# надрукувати параметри, які дають найкращі результати
print('Параметри, які дають найкращі результати :','\n\n', (grid_search.best_params_))

# надрукувати оцінювач, обраний GridSearch
print('\n\nОцінювач, обраний пошуком :','\n\n', (grid_search.best_estimator_))
```

```python
# розрахувати оцінку GridSearch CV на випробувальному наборі

print('Оцінка GridSearch CV на випробувальному наборі: {0:0.4f}'.format(grid_search.score(X_test, y_test)))
```
